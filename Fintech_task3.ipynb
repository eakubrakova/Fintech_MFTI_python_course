```python
# Импортируем библиотеки для работы с массивами и рисования графиков
import numpy as np
import matplotlib.pyplot as plt

# Задаем параметры для генерации случайных точек
N = 10 # Количество точек
a = 0.5 # Коэффициент наклона прямой
b = 0 # Коэффициент смещения прямой

# Генерируем массивы с координатами точек
x = np.arange(1, N+1) # x = [1, 2, ..., N]
y = np.random.normal(a*x + b, 1) # y = a*x + b + случайный шум с нормальным распределением

# Определяем функцию, возвращающую коэффициенты a* и b* наилучшей прямой по методу линейной регрессии
def linear_regression(x, y):
    # Вычисляем суммы квадратов координат и произведений координат
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x*y)
    sum_xx = sum(x*x)

    # Вычисляем коэффициенты a* и b*
    a_star = (N*sum_xy - sum_x*sum_y) / (N*sum_xx - sum_x**2)
    b_star = (sum_y - a_star*sum_x) / N

    return a_star, b_star

# Вычисляем коэффициенты a* и b*
a_star, b_star = linear_regression(x, y)

# Создаем массив с координатами прямой, полученной методом линейной регрессии
y_pred = a_star*x + b_star

# Рисуем график
plt.plot(x, y, 'o', label='Случайные точки')
plt.plot(x, a*x + b, label='Истинная прямая')
plt.plot(x, y_pred, label='Прямая, полученная линейной регрессией')
plt.legend()
plt.show()

# Выводим коэффициенты a* и b*
display("Коэффициент наклона прямой: ", a_star)
display("Коэффициент смещения прямой: ", b_star)

```
